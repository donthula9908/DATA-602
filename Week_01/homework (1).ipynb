## Question 1. We need to know what's included in this data.
- How many countries are present?  
- What's the timespan of the data?  
- Does each country have a row for every year present?  
- Which countries are missing years?  
- How many rows have at least 1 non-null values across the columns 3-57?

````python
# How many countries are present?
unique_countries = data['country'].unique()

a= len(unique_countries)
a
````

````python
## What's the timespan of the data?
min_year = data['year'].min()
max_year = data['year'].max()
timespan = max_year - min_year + 1
timespan
````

````python
## Does each country have a row for every year present?
unique_countries = data['country'].unique()
unique_years = data['year'].unique()
result = True
for country in unique_countries:
    country_data = data[data['country'] == country]
country_years = country_data['year'].unique()
if len(country_years) != len(unique_years):
        result = False
print(result)
````

````python
## Which countries are missing years?
unique_countries = data['country'].unique()
unique_years = data['year'].unique()
missing_years_countries = []
for country in unique_countries:
    country_data = data[data['country'] == country]
    country_years = country_data['year'].unique()
    if len(country_years) != len(unique_years):
        missing_years_countries.append(country)
missing_years_countries
````

````python
## How many rows have at least 1 non-null values across the columns 3-57?
columns_of_interest = data.iloc[:, 2:58]
non_null_count = columns_of_interest.notnull().sum(axis=1)
result = non_null_count[non_null_count > 0].count()
result
````

## Question 3: Create a graph that shows when countries started to report TB cases.

````python
# Code
country_year = data[['country', 'year']]
country_year.dropna(inplace=True)
country_year['id'] = country_year['country'].astype('category').cat.codes
first_year = country_year.groupby(['id', 'country']).first().reset_index()
plt.scatter(first_year['id'], first_year['year'])
plt.xticks(first_year['id'], first_year['country'], rotation=90)
plt.ylabel("year")
plt.show()
````
